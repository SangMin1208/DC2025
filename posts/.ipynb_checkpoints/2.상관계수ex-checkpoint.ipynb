`-` 상관계수

- X,Y에 상수를 곱해줘도 그대로, 변하지 않음

- 기울기만 변함

```r
x = rnorm(100)
y = rnorm(100) + 2*x
plot(x,y)
cor(x,y)
points(2*x, 0.25*y, col='red') 
cor(2*x, 0.25*y)
```

`-` 상관계수

- aX, bY 의 상관계수에서 aXb <0 이면 
- Corr(aX,bY) = -Corr(X,Y)
- 상관계수 크기는 그대로
- 기울기만 변함

```r
x = rnorm(100)
y = rnorm(100) + 2*x
plot(x,y)
cor(x,y)
points(-2*x, 0.1*y, col='red') 
cor(-2*x, 0.1*y)
```

`-` 상관계수 : 기울기와 관련 X, 직선에 얼마나 몰려있는가
