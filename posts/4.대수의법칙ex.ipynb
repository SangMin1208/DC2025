`-` 모평균이 3이므로 3으로 가까이 수렴하는 형태

* 분포 종류 상관 x
* 대수의 법칙

```r
rs=rep(0,1000)
for(k in 1:1000)
    {
    rs[k]=mean(rpois(1000*k,3))

    if(k%%100 == 0) print(k)
    }
```

```r
plot(rs)
```

`-` 표본의 크기가 커지면 더 가늘어짐

```r
rs=rep(0,1000)
for(k in 1:1000)
    {
    rs[k]=mean(rpois(2000*k,3))

    if(k%%100 == 0) print(k)
    }
```
